```python
# Save camera parameters

import open3d as o3d
pcd = o3d.io.read_point_cloud("/home/zichen/Downloads/Datasets_Unpack/ShapeNetCompletion/train/partial/02691156/10155655850468db78d106ce0a280f87/00.pcd")

vis = o3d.visualization.Visualizer()
vis.create_window(width=400, height=400)
vis.add_geometry(pcd)
vis.run()  # user changes the view and press "q" to terminate
vis.get_view_control().set_zoom(0.5)   # zoom in
param = vis.get_view_control().convert_to_pinhole_camera_parameters()   # get camera parameters
o3d.io.write_pinhole_camera_parameters("PCN_Cam_Parm_save.json", param)
vis.destroy_window()
```

```python
# Load the camera parameters and render the point cloud
import open3d as o3d
pcd = o3d.io.read_point_cloud("/home/zichen/Downloads/Datasets_Unpack/ShapeNetCompletion/train/complete/02691156/10155655850468db78d106ce0a280f87.pcd")

vis = o3d.visualization.Visualizer()
vis.create_window(width=400, height=400)
ctr = vis.get_view_control()
param = o3d.io.read_pinhole_camera_parameters("PCN_Cam_Parm_save.json")
vis.add_geometry(pcd)
ctr.convert_from_pinhole_camera_parameters(param)
vis.run()
vis.capture_screen_image("PCN_Train_Comp_0269_10155_Comp.png", do_render=True)
vis.destroy_window()
```
